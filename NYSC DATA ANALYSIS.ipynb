__Import all needed packages from libraries__

```python

import pandas as pd
import numpy  as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

__...extracting the data from spreadsheet__

```python
nysc= pd.read_excel('NYSC data analysis.xlsx',sheet_name= 'Sheet1',parse_dates=True)
```

```python
nysc.drop(columns='total_patient/day',inplace=True)
nysc.drop(columns='Unnamed: 8',inplace=True)
```

##### __Summary of the columns__

•	__URTI__ – Upper Respiratory Tract Infection which includes patients with signs and symptoms with Rhinitis, Rhinosinusitis,Pharyngitis among others

•	__Malaria__

•	__Gastroenteritis – related__ – this include patient with acute gastroenteritis, diarrhea, dysentery symptoms and signs.

•	__PUD__ – related – this include patient with symptoms and signs of Acute Gastritis, Peptic Ulcer Disease, GERD.

•	__Body Pain__: this includes patient with muscle cramps, sprain.

.   __Others__: includes all other disease condition apart from above-listed


```python
nysc.info()
```

__...table showing the statistics of each column__

```python
nysc.describe()
```

__... melting the nysc data into 3 columns by categorizing disease conditions__

```python
nysc_melt = pd.melt(nysc, 'Date', var_name='Disease', 
              value_name='Number_of_patients')
```

```python
nysc_melt.head(3)
```

```python
sns.set(rc={"figure.figsize":(15, 4)})
sns.lineplot(data=nysc_melt, x='Date', y='Number_of_patients', hue='Disease')
```

```python
# Create a grid : initialize it
g = sns.FacetGrid(data=nysc_melt, col='Disease', hue='Disease', col_wrap=3, )

g = g.map(plt.plot, 'Date', 'Number_of_patients')

g = g.map(plt.fill_between, 'Date', 'Number_of_patients', alpha=0.2).set_titles("{col_name} Disease")
 
g = g.set_titles("{col_name}")
 
plt.subplots_adjust(top=0.92)
g = g.fig.suptitle('Graph showing Trend of each Disease Conditions')

plt.show()
```

```python
total_patient_per_day = nysc_melt.groupby(by='Date').sum()
total_patient_per_day.head()
```

__FREQUENCE OF DISEASE PER DAY__

23rd of MArch has the highest number of patient visiting the clinic which is was the day of the swearing in of corp members, found to be the most stressful day in camp.

```python
sns.set(rc={"figure.figsize":(15, 4)})
plot=sns.lineplot(data=total_patient_per_day, x='Date', y='Number_of_patients')
```

```python
nysc.head(2)
```

__...adding weekdays and numbers of the days in the year(2022) columns to nysc data__

```python
nysc['Days']=nysc.Date.dt.day_name()
nysc['Day_date']=nysc.Date.dt.day_of_year
nysc.head()
```

__...creating a table showing total number of patients seen per disease condition between 18th of March to 4th April__

```python
patient_per_case = nysc.sum().reset_index()
patient_per_case.rename(columns= {'index':'disease', 0:'total patient per case'}, inplace=True)
patient_per_case.drop(index=[6,7],inplace=True)
```

```python
patient_per_case[['total patient per case']].astype(dtype='int64')
```

...__sorting in descending order, URTI is found to have the highest burden followed by Malaria and body pain__

```python
nysc_total = patient_per_case.sort_values(by = 'total patient per case', ascending=False)
```

__...Below is a graph showing total number of case per disease burden__

```python
sns.barplot(y='disease', x='total patient per case', data=nysc_total)
plt.title('Total Number of patient per Disease Condition')
```

```python
nysc['Days_division']= np.where(nysc['Day_date'] > nysc['Day_date'].mean(), 'second half', 'first half')
```

```python
division=nysc.groupby(by='Days_division').mean().round(0).drop(columns='Day_date')
division
```

```python
division.plot.box()
```

- __Below bar graph shows that URTI took over in prevanlency during the second half of the our stay in the camp doubling the disease next to it i.e Malaria which in itself and body pains were of highest burden in the first half of the camping.__


- __In addition, it can be noticed that body pain which seems to be among the highest 2 prevailing diseases seems to rank among the 2 lowest in the second half. this is suspect to be due to readjustment of body physiology of copy members to the  physical activities.__

```python
division.plot.bar()
plt.title('Bar Graph of the halves of days of camping and the disease burden')
```

#### INSIGHT

*  The prevalence of upper respiratory tract infection was confirmed

*  The correlation between the PUD_related cases and Body pain in tandem with stressful physical activities undergone by camp members.

*  The increase in the number of cases per day on the swearing in day


#### OTHER OBSERVATIONS

- There was more exposure to airborne particulates as camp activities (such as rehearsals for upcoming swearing in and the day itself) progresses.
- There was also a change in weather condition (cold and rainy day) on the day that URTI spiked.
- Communicable diseases easily spread due to unavoidable contacts during activities and in the lodge.
- There was a usual spurt of patient frequenting clinic just before and after severe activities or intense drill or punishments. 
- Emergencies are highest during hours of physically demanding activities.

#### RECOMMENDATIONS

1. Epidemics of the common disease listed above in camps should be foreseen and prepared for by the health care providers.

2. Seasonal variation in the burden of disease will be helpful in supply of medications and equipment.

3. In the advent of suspicion of prevalence of certain disease condition, it should be reported to appropriate quarters for immediate intervention and prevention of further spread.

4. Camp clinic staff should always be at alert during rigorous camp activities such as parades, sports, and swearing-in days.

5. There is a need for careful monitoring, continuous evaluation and rapid response in case of urgency or emergency of epidemics.

6. Common diseases with their seasonal variation should also be anticipated.

7. Every hands must always be on deck and uninterrupted flow of information among the a head of administration, health care provider and camp clinic staffs.

8. Need for thorough documentation and record-keeping cannot be overemphasized as this  will help  in data-driven evaluation of management.
